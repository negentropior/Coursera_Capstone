I collected data from different sources already, but I'd like to elaborate more about how I got "Monthly rent persquare meter". I have collected 4249 apartments' rent information of all 16 districts in Shanghai and stored it at local file "rent_info.xlsx".  In the cells down below I will transform this tabulation to get the feature "Monthly rent persquare meter".

```python
# load rent information data and view it:

rent = pd.read_excel('rent_cost.xlsx')
```

```python
rent.head(5)
```

```python
rent.shape
```

```python
# transfrom the rent dat to get the feature:
rent_ = rent.drop(rent.loc[~rent['size'].str.contains('㎡')].index, axis=0)
rent_['size'] = rent_['size'].str.replace('㎡','')
for i in rent_['size'].index:
    if rent_['size'][i].find('-') != -1:      
        rent_['size'][i] = (float(rent_['size'][i].split('-')[0])+float(rent_['size'][i].split('-')[1]))/2
rent_['size'] = rent_['size'].astype(float)
for i in rent_['price'].index:
    if rent_['price'].astype(str)[i].find('-') != -1:      
        rent_['price'][i] = (float(rent_['price'][i].split('-')[0])+float(rent_['price'][i].split('-')[1]))/2
rent_['price'] = rent_['price'].astype(float)
rent_info = rent_.groupby('borough')[['size','price']].sum()
rent_info = rent_info.reset_index()

# Add a column 'monthly_rent_persquare_meter' which is the factor I need:
rent_info['monthly_rent_persquare_meter'] = rent_info['price']/rent_info['size']

rent_info.borough = rent_info.borough.str.title()
rent_info = rent_info.merge(districts_coordinate[['District', 'District_chi']],
                            left_on='borough', right_on='District', how='left')
rent_info.District_chi[2] = '崇明区'
```

Obtain "Number of popular venues nearby" feature, name it "venue_counts":

```python
CLIENT_ID = 'NFDIPQEE1AU3AKW0K24MB5YLNF4EX3S43YLSPWZFQBDCMU0D' # your Foursquare ID
CLIENT_SECRET = 'O00B4DHU1XTEJQJ4UOPNP0ODB53TQA2SQ4RFLVVWTZNWYWXE' # your Foursquare Secret
VERSION = '20180605'
LIMIT = 100

venues_df = pd.DataFrame(columns=['District','venue_counts','venue_catagories'])
LIMIT = 500
radius = 2000

for  i in districts_coordinate.District.index:
    district_latitude = districts_coordinate.latitude[i]
    district_longitude = districts_coordinate.longitude[i]
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        CLIENT_ID, 
        CLIENT_SECRET, 
        VERSION, 
        district_latitude, 
        district_longitude, 
        radius, 
        LIMIT)
    results = requests.get(url).json()
    v_counts = results['response']['totalResults']
    li = []
    if v_counts > 100:
        lim = 100
    else:
        lim = v_counts
    for j in range(lim):
        li.append(results['response']['groups'][0]['items'][j]['venue']['categories'][0]['name'])
    set_li = set(li)
    dic = {}
    for item in set_li:
        dic[item] = li.count(item)
    d_order = sorted(dic.items(), key=lambda x:x[1], reverse=True)
    v_catagories = d_order
    row = {"District": districts_coordinate.District[i], "venue_counts": v_counts, "venue_catagories": v_catagories}
    venues_df = venues_df.append(row, ignore_index=True)
```

## 4.Clustering districts in Shanghai

Now let's introduce K-Means to cluster our districts into different groups:

```python
# step1: using Elbow method to figure out the best "k"

from sklearn.cluster import KMeans
df = pd.DataFrame(feature_mtx, index=data_plot.index, columns=data_plot.columns)
df_features = df 
SSE = []  
for k in range(1,10):
    estimator = KMeans(n_clusters=k) 
    estimator.fit(df_features)
    SSE.append(estimator.inertia_) 
X = range(1,10)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X,SSE,'o-')
plt.show()
```

From the chart above, we can figure out the best k is when k=3.

```python
# step2: Clustering feature_mtx and view results

kclusters =3
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(feature_mtx)
kmeans.labels_
```

```python
# step3: Insert labels into features dataframe and check each clusterings

df_check = df.copy()
df_check.insert(0, 'Cluster Labels', kmeans.labels_)
```

## 5.Naming each clusterings based on difference of features and plotting them

From above chart, we can name each clusterings as follow:
    
1.Cluster Labels=0 --> 'low rent, low population density, less venues'  
2.Cluster Labels=1 --> 'high rent, high population density, more venues'  
3.Cluster Labels=2 --> 'medium rent, medium population density, medium venues'

```python
# Plotting clusterings in map

df_check['marker'] = ''
for i in range(len(df_check)):
    if df_check['Cluster Labels'][i] == 0:
        df_check['marker'][i] = df_check.index[i]+'-'+'low rent, low population density, less venues'
    elif df_check['Cluster Labels'][i] == 1:
        df_check['marker'][i] = df_check.index[i]+'-'+'high rent, high population density, more venues'
    else:
        df_check['marker'][i] = df_check.index[i]+'-'+'medium rent, medium population density, medium venues'

clus_plot = districts_coordinate.merge(df_check.reset_index()[['District','marker']],
                                        on='District',how='left')

clus_plot
```

```python
# add markers to map
for lat, lng, label in zip(clus_plot['latitude'], clus_plot['longitude'], clus_plot['marker']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(shanghai_map)  
    
shanghai_map
```
