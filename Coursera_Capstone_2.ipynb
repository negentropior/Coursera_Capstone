```python
import pandas as pd

df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

df_Tor = pd.DataFrame(df[0])
df_Tor = df_Tor.loc[df_Tor.Borough != 'Not assigned']
df_Tor = df_Tor.reset_index(drop=True)

df_Tor.loc[(df_Tor.Neighbourhood == 'Not assigned'),'Neighbourhood'] =\
df_Tor.loc[(df_Tor.Neighbourhood == 'Not assigned'),'Borough']
df_tor = pd.DataFrame(columns=df_Tor.columns)

for i in range(len(df_Tor)):
    j = len(df_tor)
    if df_Tor['Postal Code'][i] not in set(df_tor['Postal Code']):
        a = {"Postal Code":df_Tor['Postal Code'][i],"Borough":df_Tor['Borough'][i],"Neighbourhood":df_Tor['Neighbourhood'][i]}
        df_tor = df_tor.append(a, ignore_index=True)       
    elif df_Tor['Postal Code'][i] in set(df_tor['Postal Code']):
        df_tor['Neighbourhood'][df_Tor.loc[df_tor['Postal Code'] == df_Tor['Postal Code'][i]].index] = \
            df_tor['Neighbourhood'][df_Tor.loc[df_tor['Postal Code'] == df_Tor['Postal Code'][i]].index] +',' + df_Tor['Neighbourhood'][i]
```

```python
import requests
from geopy.geocoders import Nominatim
import folium
```

```python

address = 'Toronto'
geolocator = Nominatim(user_agent="yto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_tor['Latitude'], df_tor['Longitude'], df_tor['Borough'], df_tor['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
